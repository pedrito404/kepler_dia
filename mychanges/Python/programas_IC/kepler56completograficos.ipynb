```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import shutil
```

```python
a = !find Kepler56_modelo1/Kepler56bcd/follow*
b = !find Kepler56_modelo2/Kepler56bcd/follow* 

k56_1_b = pd.read_csv(a[0],delim_whitespace=True,header =None,engine='c')
k56_1_b[0] = k56_1_b[0]/365.25
k56_1_c = pd.read_csv(a[1],delim_whitespace=True,header =None,engine='c')
k56_1_c[0] = k56_1_c[0]/365.25
k56_1_d = pd.read_csv(a[2],delim_whitespace=True,header =None,engine='c')
k56_1_d[0]=k56_1_d[0]/365.25

k56_2_b = pd.read_csv(b[0],delim_whitespace=True,header =None,engine='c')
k56_2_b[0] = k56_2_b[0]/365.25
k56_2_c = pd.read_csv(b[1],delim_whitespace=True,header =None,engine='c')
k56_2_c[0] = k56_2_c[0]/365.25
k56_2_d = pd.read_csv(b[2],delim_whitespace=True,header =None,engine='c')
k56_2_d[0] = k56_2_d[0]/365.25

k56_1_b.columns=['t','a','e','i','capom','omega','capm','peri','apo','obar']
k56_1_c.columns=['t','a','e','i','capom','omega','capm','peri','apo','obar']
k56_1_d.columns=['t','a','e','i','capom','omega','capm','peri','apo','obar']
k56_2_b.columns=['t','a','e','i','capom','omega','capm','peri','apo','obar']
k56_2_c.columns=['t','a','e','i','capom','omega','capm','peri','apo','obar']
k56_2_d.columns=['t','a','e','i','capom','omega','capm','peri','apo','obar']
```

```python
plt.figure(figsize=(10,5))
plt.plot(k56_1_b['t'],k56_1_b['a'],color='red',label='Kepler56-b')
plt.plot(k56_1_c['t'],k56_1_c['a'],color='blue',label='Kepler56-c')
plt.plot(k56_1_d['t'],k56_1_d['a'],color='yellow',label='Kepler56-d')
plt.xlabel('Tempo (Anos)')
plt.ylabel('Semi-eixo maior (UA)')
plt.title('Kepler-56 - Modelo 1')
plt.legend()
plt.savefig('k56_a.pdf')
plt.show()

plt.figure(figsize=(10,5))
plt.plot(k56_1_b['t'],k56_1_b['e'],color='red',label='Kepler56-b')
plt.plot(k56_1_c['t'],k56_1_c['e'],color='blue',label='Kepler56-c')
plt.plot(k56_1_d['t'],k56_1_d['e'],color='yellow',label='Kepler56-d')
plt.xlabel('Tempo (Anos)')
plt.ylabel('Excentricidade')
plt.title('Kepler-56 - Modelo 1')
plt.legend()
plt.savefig('k56_e.pdf')
plt.show()

plt.figure(figsize=(10,5))
plt.plot(k56_1_b['t'],k56_1_b['i'],color='red',label='Kepler56-b')
plt.plot(k56_1_c['t'],k56_1_c['i'],color='blue',label='Kepler56-c')
plt.plot(k56_1_d['t'],k56_1_d['i'],color='yellow',label='Kepler56-d')
plt.xlabel('Tempo (Anos)')
plt.ylabel('Inclinao ($^{\circ}$)')
plt.title('Kepler-56 - Modelo 1')
plt.legend()
plt.savefig('k56_i.pdf')
plt.show()
```
